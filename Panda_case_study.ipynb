1) Import pandas, numpy


```python
import pandas as pd
import numpy as np
```


2) Load dataset using pd.read_csv()

```python
data = pd.read_csv('/content/IMDB-Movie-Data.txt')
data
```

```python
data.info()
```

3) Display: shape, head, column names, datatypes, missing value

```python
print(data.shape)
print(data.head())
print(data.columns)
print(data.dtypes)
print(data.isnull().sum())
```

4) Check and remove duplicates

```python
print("Duplicate rows before removal:", data.duplicated().sum())
data = data.drop_duplicates()
```

5) Rename columns ("Title" > "Movie_Title")

```python
data = data.rename(columns={"Title": "Movie_Title"})
print("Updated Columns:", data.columns.tolist())
```

6) Get unique genres


```python
unique_genres = data['Genre'].str.split(',').explode().unique()
print("Unique Genres:\n", unique_genres)
```

7) Extract number of genres per movie


```python
data['Num_Genres'] = data['Genre'].apply(lambda x: len(x.split(',')))
print("\nNumber of Genres per Movie:\n", data[['Movie_Title', 'Num_Genres']].head())
```

8) Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
data['Revenue (Millions)'] = data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].mean())
data['Metascore'] = data['Metascore'].fillna(data['Metascore'].mean())

print("Missing values after handling:\n", data[['Revenue (Millions)', 'Metascore']].isnull().sum())
```

9) Top 5 movies by Rating

```python
top_5_rating = data.nlargest(5, 'Rating')[['Movie_Title', 'Rating']]
print("Top 5 Movies by Rating:\n", top_5_rating)
```

10) Movies with highest/lowest revenue


```python
highest_revenue = data.loc[data['Revenue (Millions)'].idxmax()]
lowest_revenue = data.loc[data['Revenue (Millions)'].idxmin()]
print("\nHighest Revenue Movie:", highest_revenue['Movie_Title'], "-", highest_revenue['Revenue (Millions)'])
print("Lowest Revenue Movie:", lowest_revenue['Movie_Title'], "-", lowest_revenue['Revenue (Millions)'])
```

11) Avg. rating by Director

```python
avg_rating_director = data.groupby('Director')['Rating'].mean()
print("Avg Rating by Director (First 5):\n", avg_rating_director.head())
```

12) Movies per year

```python
movies_per_year = data['Year'].value_counts().sort_index()
print("\nMovies Per Year:\n", movies_per_year)
```

13) Movies with rating > 8 & revenue > 100M

```python
top_tier_movies = data[(data['Rating'] > 8) & (data['Revenue (Millions)'] > 100)]
print("Rating > 8 and Revenue > 100M:\n", top_tier_movies[['Movie_Title', 'Rating', 'Revenue (Millions)']])
```

14) Group by Director: avg Rating, Revenue

```python
director_stats = data.groupby('Director').agg({
    'Rating': 'mean',
    'Revenue (Millions)': 'mean'
})
print("Director Stats (First 5):\n", director_stats.head())
```

15) Group by Year: count, avg Rating

```python
year_status = data.groupby('Year').agg({
    'Movie_Title': 'count',
    'Rating':'mean'
}).rename(columns={'Movie_Tilte': 'Movie_Count'})
print("\nYearly Status:\n", year_status)
```

16) Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies


```python
data['Profitability'] = data['Revenue (Millions)'] / data['Runtime (Minutes)']
top_10_profitable = data.nlargest(10, 'Profitability')[['Movie_Title', 'Profitability']]
print("Top 10 Profitable Movies:\n", top_10_profitable)
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 ≤ Rating < 7.5: Average
c) Rating ≥ 7.5: Excellent

```python
def categorize_rating(rating):
    if rating < 6:
        return 'Poor'
    elif 6 <= rating < 7.5:
        return 'Average'
    else:
        return 'Excellent'
data['Rating_Category'] = data['Rating'].apply(categorize_rating)
print("Rating Categories:\n", data[['Movie_Title', 'Rating', 'Rating_Category']].head())
```

18) Create a custom function to score movies based on:
i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def calculate_score(row):
    return (row['Rating'] * 10) + row['Revenue (Millions)']
data['Custom_Score'] = data.apply(calculate_score, axis=1)

print("Custom Scores:\n", data[['Movie_Title', 'Rating', 'Revenue (Millions)', 'Custom_Score']].head())
```
