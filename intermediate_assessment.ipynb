```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    try:
        with open(filename, "r") as f:
            content = f.read().lower()


        words = re.findall(r'\w+', content)


        counts = Counter(words)


        sorted_words = sorted(counts.items(), key=lambda x: (-x[1], x[0]))


        return [word for word, count in sorted_words[:k]]

    except FileNotFoundError:
        return []


with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")


print("Top 3 words:", top_k_words("sample.txt", 3))
```

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for i in range(n):

        user_input = input(f"Transaction {i+1} - Enter category and amount (e.g., Food 200): ")
        category, amount = user_input.split()
        amount = float(amount)


        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount


    total = sum(expenses.values())


    highest_category = max(expenses, key=expenses.get) if expenses else None

    print("\n--- Results ---")
    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise Total:", expenses)


analyze_expenses()
```

```python
def clean_data():
    n = int(input("Enter number of values you want to enter: "))

    data = []

    for i in range(n):
        value = input(f"Enter value {i+1} (can be text, positive, or negative number): ")

        try:
            num = float(value)


            if num >= 0:
                data.append(num)

        except ValueError:

            pass


    if data:
        avg = sum(data) / len(data)
        max_val = max(data)
    else:
        avg = 0
        max_val = None

    print("\n--- Results ---")
    print("Clean Data List:", data)
    print("Average:", avg)
    print("Max Value:", max_val)

clean_data()
```
