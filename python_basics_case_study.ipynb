Create Account Information

```python

account_name = "Ashik"
account_number = 123456
balance = 5000.0

print("Account Holder:", account_name)
print("Account Number:", account_number)
print("Current Balance:", balance)
```

```python
def deposit(balance, amount):
    balance = balance + amount
    print("Deposited Amount:", amount)
    print("Updated Balance:", balance)
    return balance

balance = deposit(balance, 2000)
```

```python
def withdraw(balance, amount):
    if amount <= balance:
        balance = balance - amount
        print("Withdrawal Successful")
        print("Withdrawn Amount:", amount)
        print("Remaining Balance:", balance)
    else:
        print("Insufficient Balance")

    return balance

balance = withdraw(balance, 3000)
```

```python
def check_balance(balance):
    print("Available Balance:", balance)

check_balance(balance)
```

Weather Data Analyzer

```python
temperature = [30, 32, 31, 29, 33]
humidity = [70, 65, 75, 80, 68]

print("Temperature Data:", temperature)
print("Humidity Data:", humidity)
```

```python
def average_temperature(temp):
    avg = sum(temp) / len(temp)
    print("Average Temperature:", avg)

average_temperature(temperature)
```

```python
def average_humidity(hum):
    avg = sum(hum) / len(hum)
    print("Average Humidity:", avg)

average_humidity(humidity)
```

```python
current_temperature = 33
current_humidity = 68

if current_temperature > 32:
    print("Weather Condition: Sunny")
elif current_humidity > 75:
    print("Weather Condition: Rainy")
else:
    print("Weather Condition: Cloudy")
```
