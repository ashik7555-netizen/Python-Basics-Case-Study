Rows: Check for duplicate rows in the dataset, if any, and remove them.
Columns: Identify and drop duplicate columns, if any, based on their values.

```python
duplicate_columns = df.columns[df.T.duplicated()].tolist()
if duplicate_columns:
    print(f"Found duplicate columns: {duplicate_columns}. Dropping them.")
    df = df.drop(columns=duplicate_columns)
else:
    print("No duplicate columns found.")
```

```python
display(df.head())
```

```python
columns_to_drop_by_value = []
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        col1 = df.columns[i]
        col2 = df.columns[j]
        if df[col1].equals(df[col2]):
            columns_to_drop_by_value.append(col2)

if columns_to_drop_by_value:
    print(f"Found columns with identical values: {columns_to_drop_by_value}. Dropping them.")
    df = df.drop(columns=columns_to_drop_by_value)
else:
    print("No columns with identical values found.")

display(df.head())
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:")
display(missing_values[missing_values > 0].sort_values(ascending=False))
```

```python
numerical_cols_with_missing = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols_with_missing = df.select_dtypes(include=['object']).columns

if 'missing_values' not in globals():
    missing_values = df.isnull().sum()

for col in missing_values[missing_values > 0].index:
    if col in numerical_cols_with_missing:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Imputed numerical column '{col}' with median: {median_val}")
    elif col in categorical_cols_with_missing:
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
        print(f"Imputed categorical column '{col}' with mode: {mode_val}")

print("\nMissing values after imputation:")
display(df.isnull().sum()[df.isnull().sum() > 0])
display(df.head())
```

```python
print('Column Data Types:')
display(df.dtypes)

print('\nMissing Value Imputation Strategy:')

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

for col_name, missing_count in missing_values[missing_values > 0].items():
    if col_name in numerical_cols:
        print(f"- '{col_name}' (Numerical): Imputed with the median value. (Originally {missing_count} missing values)")
    elif col_name in categorical_cols:
        print(f"- '{col_name}' (Categorical): Imputed with the mode value. (Originally {missing_count} missing values)")
    else:
        print(f"- '{col_name}': Data type not clearly numerical or categorical for imputation strategy. (Originally {missing_count} missing values)")
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Sale Price' in numerical_columns:
    numerical_columns.remove('Sale Price')

print("Numerical columns (excluding 'Sale Price'):")
print(numerical_columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print("DataFrame after Min-Max scaling of numerical columns:")
display(df.head())
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

print("Categorical columns:")
print(categorical_columns)
```

```python
print("Unique values for 'Waterfront View':")
print(df['Waterfront View'].unique())

print("\nUnique values for 'No of Times Visited':")
print(df['No of Times Visited'].unique())

print("\nUnique values for 'Condition of the House':")
print(df['Condition of the House'].unique())

print("\nUnique values for 'Date House was Sold' (first 10):")
print(df['Date House was Sold'].unique()[:10])
```

```python
df = pd.get_dummies(df, columns=['Waterfront View'], drop_first=True, dtype=int)

visit_order = ['Once', 'Twice', 'Thrice', 'Four']
df['No of Times Visited'] = df['No of Times Visited'].map({val: i for i, val in enumerate(visit_order)})

condition_order = ['Bad', 'Okay', 'Fair', 'Good', 'Excellent']
df['Condition of the House'] = df['Condition of the House'].map({val: i for i, val in enumerate(condition_order)})

df['Date House was Sold'] = pd.to_datetime(df['Date House was Sold'])
df['Sale_Year'] = df['Date House was Sold'].dt.year
df['Sale_Month'] = df['Date House was Sold'].dt.month

df = df.drop(columns=['Date House was Sold'])

print("DataFrame after applying encoding techniques and feature engineering:")
display(df.head())
```

```python
original_rows = df.shape[0]

print(f"Original number of rows: {original_rows}")

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

outlier_columns_to_process = [
    'No of Bedrooms',
    'No of Bathrooms',
    'Flat Area (in Sqft)',
    'Lot Area (in Sqft)',
    'Overall Grade',
    'Area of the House from Basement (in Sqft)',
    'Basement Area (in Sqft)',
    'Renovated Year',
    'Latitude',
    'Longitude',
    'Living Area after Renovation (in Sqft)',
    'Lot Area after Renovation (in Sqft)'
]


for col in outlier_columns_to_process:
    if col in df.columns:
        df = remove_outliers_iqr(df, col)
        print(f"Outliers removed from '{col}'. Current shape: {df.shape}")

final_rows = df.shape[0]
print(f"\nNumber of rows after outlier removal: {final_rows}")
print(f"Number of rows removed: {original_rows - final_rows}")
display(df.head())
```

```python
print("Outlier Detection using IQR Method:")

for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    if not outliers.empty:
        print(f"\nColumn: '{col}'")
        print(f"  Q1: {Q1:.2f}")
        print(f"  Q3: {Q3:.2f}")
        print(f"  IQR: {IQR:.2f}")
        print(f"  Lower Bound: {lower_bound:.2f}")
        print(f"  Upper Bound: {upper_bound:.2f}")
        print(f"  Number of outliers: {len(outliers)}")

    else:
        print(f"\nColumn: '{col}' - No significant outliers detected based on IQR.")
```
