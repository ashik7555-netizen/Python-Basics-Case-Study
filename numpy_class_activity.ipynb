Numpy

```python
import numpy as np

arr = np.arange(10, 50)
matrix = arr.reshape(4, 10)
last_col = matrix[:, -1]

print("Reshaped Matrix:\n", matrix)
print("Last Column:", last_col)
```

```python
m1 = np.full((3,3),5)
m2 = np.arange(1,10).reshape(3,3)
added = m1 + m2
multiplied = m1 + m2
transposed = multiplied.T
print("Transposed Result:\n", transposed)
```

```python
np.random.seed(42)
rand_arr = np.random.randint(10, 101, 15)
print("Original:", rand_arr)

mask = rand_arr > 50
replaced_count = np.sum(mask)
rand_arr[mask] = -1

print("Modified:", rand_arr)
print("Number of replaced values:", replaced_count)
```

```python
np.random.seed(42)
float_arr = np.random.rand(5, 4)
row_sums = np.sum(float_arr, axis=1)
max_sum_row_idx = np.argmax(row_sums)
print("Row index with max sum:", max_sum_row_idx)
row_mins = float_arr.min(axis=1, keepdims=True)
row_maxs = float_arr.max(axis=1, keepdims=True)
normalized_arr = (float_arr - row_mins) / (row_maxs - row_mins)

print("Normalized Array:\n", np.round(normalized_arr, 2))
```

```python
arr5 = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_elements, counts = np.unique(arr5, return_counts=True)
duplicates = unique_elements[counts > 1]

print("Unique elements:", unique_elements)
print("Duplicates:", duplicates)
```

```python
arr6 = np.arange(1, 37).reshape(6, 6)
submatrix = arr6[1:5, 1:5]
evens = submatrix[submatrix % 2 == 0]

print("Middle Submatrix:\n", submatrix)
print("Even numbers:", evens)
```

```python
arr7 = np.arange(12).reshape(3, 4)
reshaped_2x6 = arr7.reshape(2, 6)
flattened = reshaped_2x6.flatten()
final_shape = flattened.reshape(4, 3)

print("Final 4x3 Array:\n", final_shape)
```

```python
a = np.array([[1, 2], [3, 4]])
a = np.append(a, [[5, 6]], axis=0)
col_to_append = np.array([[7], [8], [9]])
a = np.append(a, col_to_append, axis=1)

print("Final Appended Array:\n", a)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
joined = np.concatenate((a, b, c))
matrix3x3 = joined.reshape(3, 3)
diagonal = np.diagonal(matrix3x3)

print("3x3 Matrix:\n", matrix3x3)
print("Diagonal Elements:", diagonal)
```

```python
arr10 = np.arange(1,21)
divisible_vals =arr10[(arr10 % 3 == 0) | (arr10 % 5 == 0)]
count_divisible = len(divisible_vals)
print("Divisible Values:", divisible_vals)
print("Count of Divisible Values:", count_divisible)
```

Pandas


```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/titanic_dataset (1).csv')
df
```

```python
("\n--- Q2: Missing values per column ---")
df
(df.isnull().sum())
df
```

```python
print("\n--- Q3: Survival percentage by gender ---")
survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print(survival_by_gender)
```

```python
print("\n--- Q4: Age Statistics (Average, Median, Std Dev) ---")
age_stats = df['Age'].agg(['mean', 'median', 'std'])
print(age_stats)
```

```python
print("\n--- Q5: Passenger count by class ---")
class_counts = df['Pclass'].value_counts().sort_index()
print(class_counts)
```

```python
("\n--- Q6: Female, under 18, and survived ---")
df
young_female_survivors = df[(df['Sex'] == 'female') & (df['Age'] < 18) & (df['Survived'] == 1)]
(young_female_survivors)
df
```

```python
print("\n--- Q7: Replacing missing ages with median ---")
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
print(f"Missing 'Age' values after replacement: {df['Age'].isnull().sum()}")
```

```python
print("\n--- Q8: Creating age_group column ---")
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['Age'].apply(categorize_age)
print(df[['Age', 'age_group']].head())
```

```python
print("\n--- Q9: Top 5 highest fares ---")
top_fares = df.nlargest(5, 'Fare')
print(top_fares[['Name', 'Fare']])
```

```python
print("\n--- Q10: Average fare by class and sex ---")
avg_fare_grouped = df.groupby(['Pclass', 'Sex'])['Fare'].mean()
print(avg_fare_grouped)
```

```python
print("\n--- Q11: Renaming columns ---")
df = df.rename(columns={'Sex': 'gender', 'Fare': 'ticket_fare'})
print("Current columns:", df.columns.tolist())
```

```python
print("\n--- Q12: Survival counts by class ---")
survival_counts_by_class = df.groupby(['Pclass', 'Survived']).size()
print(survival_counts_by_class)
```
